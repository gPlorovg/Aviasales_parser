```python
import pandas as pd
import requests
from itertools import product
from tqdm import tqdm

TOKEN = 'your API token'
def get_price(date, fr, to):
    date = date.split()[0]
    resp = requests.get(f'https://api.travelpayouts.com/aviasales/v3/prices_for_dates?currency=rub&origin={fr}&destination={to}&departure_at={date}&token={TOKEN}')
    
    if resp.status_code != 200:
        print('ПАМАГИТЕ')
    
    if not resp.json()['data']:
        return 0
    return resp.json()['data'][0]['price']

data = pd.date_range('2023-04-21', '2023-12-31').to_series()

to_list = ['ANK', 'AYT', 'IST']
fr_list = ['AER', 'IKT']

for fr_, to_ in tqdm(product(fr_list, to_list)):
    res = data.apply(lambda date: get_price(str(date), fr_, to_))
    res.to_csv(f'{fr_}-{to_}.csv', sep=';')
```
